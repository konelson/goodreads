```python
def book_summary(url):

    response = requests.get(url)
    page = response.text   
    soup = BeautifulSoup(page,"lxml")
    
    image = soup.find(class_="editionCover").findNext()['src']
    description = soup.find(id = 'description').text
     
    return image, description
```
